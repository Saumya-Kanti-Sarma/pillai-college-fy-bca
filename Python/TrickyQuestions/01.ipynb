### Q2 â€” Nested loops and continue
How many pairs will be printed? (See next cell)

```python
for i in range(3):
    for j in range(3):
        if i == j:
            continue
        print(i, j)
```

### Q3 â€” while with break/else
Will "Loop finished" print? (See next cell)

```python
count = 0
while count < 5:
    count += 1
    if count == 3:
        break
else:
    print("Loop finished")
print("Exited")
```
